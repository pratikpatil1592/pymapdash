```python
import folium
import pandas as pd
```

```python
seismic_data = pd.read_csv('seismic',delimiter='\t',encoding='utf-8')
seismic_data.head()
```

```python
def by_magnitude(magnitude):

    locations = seismic_data[['EQ_PRIMARY','LATITUDE','LONGITUDE','LOCATION_NAME','DAY','MONTH','YEAR']]
    
    subset = locations[locations["EQ_PRIMARY"].values == magnitude]
    
    total_count = subset.merge(locations)
    return total_count

by_magnitude(6.6)
```

```python
def plot_eq_map(total_count):
    folium_map = folium.Map(location=[31.107, 51.678],
                       zoom_start = 2,
                       tiles = "CartoDB dark_matter")
    
    for index, row in total_count.iterrows():
        
        eq_mag = (row["EQ_PRIMARY"])

        popup_text = "Magnitude: {}<br> Location: {}<br> Date: {}/{}/{}"
        popup_text = popup_text.format(row["EQ_PRIMARY"],
                                      row["LOCATION_NAME"],
                                      row["DAY"], row["MONTH"], row["YEAR"])

        radius = eq_mag*2

        if radius >= 14:
            color="#FF0000" # red
        elif radius >= 12 and radius < 14:
            color="#007849" # green
        elif radius >= 8 and radius < 12:
            color="#0A8A9F" # teal
        else:
            color="#FFCE00" # yellow            
            
        folium.CircleMarker(location = (row['LATITUDE'], 
                                        row['LONGITUDE']), 
                            radius = radius,
                            color = color,
                            popup = popup_text,
                            fill=True).add_to(folium_map)
    return folium_map
```

```python
total_count = by_magnitude(6.5)
folium_map = plot_eq_map(total_count)
folium_map
```
